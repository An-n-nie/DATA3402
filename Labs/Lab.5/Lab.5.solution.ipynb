Quiz 2: Write a function make_deck that returns a list of all of the cards in a standard card deck. The return should be a list of tuples of pairs of suit and value. For example the 10 of Clubs would be  ('Clubs', 10) and Queen of Hearts would be ('Hearts', 'Queen'). Recall that a deck has 52 cards, divided into 4 suits (Clubs, Diamonds, Hearts, and Spades), and that each suit has 13 cards: 2 to 10, Jack, Queen, King, and Ace. Summit your solution with Lab 5.


```python
def make_deck():
    suits=['Clubs', 'Diamonds','Hearts','Spades']
    values=[2,3,4,5,6,7,8,9,10,'Jack','Queen','King','Spades']
    deck= [(value,suit) for value in values for suit in suits]
    return deck
```

```python
import copy

class Matrix:
    
    def __init__(self,n,m=None):
        if isinstance(n,list):
            row_length= {len(row) for row in n}
            if len(row_length) > 1:
                raise ValueError ("Rows must have same amount of columns")

            self.matrix=n
            self.n=len(n)
            self.m=len(n[0]) if self.n > 0 else 0
        
        elif isinstance(n,int) and isinstance(m,int): 
            if n<0 or m<0:
                raise ValueError ("Dimensions must not be negative")
                                  
            self.n=n
            self.m=m
            self.matrix=[ [0 for _ in range(m)] for _ in range(n)]

        else:
            raise TypeError 
            

    def __getitem__(self,index):
        if isinstance(index,tuple):
            i,j=index
            if isinstance(i,int) and isinstance(j,int): #single pair
                return self.matrix[i][j]

            elif isinstance(i,slice) and isinstance(j,int): #for column slice - single line
                return Matrix([ [self.matrix[n][j]] for n in range(*i.indices(self.n))])

            elif isinstance(i,int) and isinstance(j,slice): #for row slice - single line
                return Matrix([self.matrix[i][j]])

            elif isinstance(i,slice) and isinstance(j,slice): #for sub matrix slice
                return Matrix([ [self.matrix[row][col] for col in range(*j.indices(self.m))] for row in range(*i.indices(self.n))])

        elif isinstance(index,slice): #support row slicing
            return Matrix(self.matrix[index])

        elif isinstance(index, int): #single row
            return self.matrix[index]

    
    def __setitem__(self,index,value):
        if isinstance(index,tuple):
            i,j=index
            self.matrix[i][j]=value
        else:
            self.matrix[index]=value

    def assign(self,other):
        if isinstance(other, Matrix):
            if (other.n,other.m) != (self.n, self.m):
                raise ValueError ("Dimensions did not match")

            self.matrix=copy.deepcopy(other.matrix) #since we can't assign, i just use deepcopy

        elif isinstance(other,list):
            row_length= {len(row) for row in other}
            if len(row_length) > 1:
                raise ValueError ("Dimensions did not match")
            if len(other) != self.n or (self.n > 0 and len(other[0]) != self.m):
                raise ValueError ("List dimension did not match")
                
            self.matrix=copy.deepcopy(other)
        else:
            raise TypeError ("Matrix or list of lists only")
        return self 

    
    def __repr__(self):
        return "\n".join(str(row) for row in self.matrix)


#Requested methods    
    def shape(self):
        return self.n, self.m

    
    def transpose(self):
        if self.n ==0 or self.m == 0:
            return Matrix([])
        
        transposed= [[self.matrix[j][i] for j in range(self.n)] for i in range(self.m)]
        return Matrix(transposed)

    
    def row(self,n):
        return Matrix([self.matrix[n]])

    
    def column(self,n):
        return Matrix([[self.matrix[i][n]] for i in range(self.n)])

    
    def to_list(self):
        return copy.deepcopy(self.matrix) if self.n > 0 else []

    
    def block(self,n_0,n_1,m_0,m_1):
        sub_matrix=[row[m_0:m_1] for row in self.matrix[n_0:n_1]]
        return Matrix(sub_matrix)


# Mathy functions    
    def equals(self,N):
        if not isinstance(N, Matrix):
            return False
        return self.matrix == N.matrix

    
    def scalarmul(self,c):
        return Matrix([[c* self.matrix[i][j] for j in range(self.m)] for i in range(self.n)])

    
    def add(self,N):
        if self.shape() != N.shape():
            raise ValueError ("Matrix dimensions did not match")
            
        return Matrix([[self.matrix[i][j] + N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])

    
    def sub(self,N):
        if self.shape() != N.shape():
            raise ValueError ("Matrix dimensions did not match")
            
        return Matrix([[self.matrix[i][j] - N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])

    
    def mat_mult(self,N):
        if self.shape()[1] != N.shape()[0]:
            raise ValueError ("Matrix dimensions did not match")
        
        return Matrix([ [sum (self.matrix[i][k] * N.matrix[k][j] for k in range(self.m)) for j in range(N.m)] for i in range(self.n)])

    
    def element_mult(self,N):
        if self.shape() != N.shape():
            raise ValueError ("Matrix dimensions did not match")
            
        return Matrix([[self.matrix[i][j] * N.matrix[i][j] for j in range(self.m)] for i in range(self.n)])
        

# Overloading    
    def __add__(self,other):
        return self.add(other)


    def __sub__(self,other):
        return self.sub(other)


    def __eq__(self,other):
        return self.equals(other)

    
    def __mul__(self,other): # will use * 
        if isinstance(other,(int,float)): #for sclar multiplication
            return self.scalarmul(other)
        elif isinstance(other, Matrix): #for element-wise multiplication
            return self.element_mult(other)
        else: 
            raise TypeError

    __rmul__=__mul__ #allow multiplication from either side

    def __matmul__(self,other): #will use @ 
        if isinstance(other, Matrix):
            return self.mat_mult(other)
        else:
            raise TypeError
```

```python
#Standalone functions

def constant(n,m,c):
    if n<0 or m<0:
        raise ValueError ("Dimensions must not be negative")
        
    c_matrix= [ [float(c) for _ in range(m)] for _ in range(n)]
    return c_matrix


def zeros(n,m):
    if n<0 or m<0:
        raise ValueError ("Dimensions must not be negative")

    z_matrix= [[0 for _ in range(m)] for _ in range(n)]
    return z_matrix


def ones(n,m):
    if n<0 or m<0:
        raise ValueError ("Dimensions must not be negative")

    one_matrix= [[1 for _ in range(m)] for _ in range(n)]
    return one_matrix


def eye(n):
    if n<0:
        raise ValueError ("Dimensions must not be negative")

    eye_matrix= [[1 if i==j else 0 for j in range(n)] for i in range(n)]
    return eye_matrix
```
